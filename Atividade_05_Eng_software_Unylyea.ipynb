### f) Vamos agora imprimir os nomes atuais das colunas, com o comando columns, e depois renomear algumas destas colunas, com o comando rename, e imprimir novamente os nomes das colunas para termos a certeza que foram alterados com o comando da forma que queríamos;


```
# Imprime os nomes atuais das colunas
print(df1.columns)

# Renomeia as colunas 'Máquina 1' e 'Máquina 2' para 'M1' e 'M2', respectivamente
df1 = df1.rename(columns={'Máquina 1': 'M1', 'Máquina 2': 'M2', 'Máquina 3': 'M3', 'Máquina 4': 'M4', 'Máquina 5': 'M5', 'Máquina 6': 'M6', 'Máquina 7': 'M7', 'Máquina 8': 'M8', 'Máquina 9': 'M9', 'Máquina 10': 'M10', 'Máquina 11': 'M11', 'Máquina 12': 'M12',})

# Imprime os novos nomes das colunas
print(df1.columns)
```

### g) Agora seguindo a apostila acrescente mais três comandos usando métodos do Pandas e anote as saídas no relatório que deverá ser enviado nesta tarefa.

```
# Ordenar o DataFrame por uma coluna específica:
# Podemos usar o método sort_values() para ordenar o DataFrame com base nos valores de uma coluna específica
df_sorted = df1.sort_values('M1')
df_sorted.head()
```

```
#Agrupar os dados por uma coluna e calcular estatísticas:
#Podemos usar o método groupby() para agrupar os dados do DataFrame com base nos valores de uma coluna específica e, em seguida, calcular estatísticas para cada grupo
df_grouped = df1.groupby('M1').agg({'M2': ['mean', 'std'], 'M3': ['mean', 'std']})
print(df_grouped.head())
```

```
#Filtrar linhas com base em uma condição:
#Podemos usar operadores lógicos para criar condições e filtrar linhas do DataFrame com base nessas condições
df_filtered = df1[df1['M1'] > 0.5]
print(df_filtered.head())
```
